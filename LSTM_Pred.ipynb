```python
# Extract the 'quote' column and store it in a new Series called 'quotes'
quotes = df['quote']
```

```python
# Display the 'quotes' Series to view its content
quotes
```

```python
# Convert all quotes to lowercase for consistent text processing
quotes = quotes.str.lower()
# Import the string module to get punctuation characters
import string
# Create a translation table to remove all punctuation from the quotes
translator = str.maketrans('', '', string.punctuation)
quotes = quotes.apply(lambda x: x.translate(translator))
```

```python
# Display the quotes after converting to lowercase and removing punctuation
quotes
```

```python
# Import the Tokenizer from TensorFlow Keras for text tokenization
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
vocab_size = 8978 # This sets the maximum number of unique words the tokenizer will keep.
# Initialize the Tokenizer with the specified vocabulary size
tokenizer = Tokenizer(num_words=vocab_size)
# Fit the tokenizer on the preprocessed quotes to build its vocabulary (word_index)
tokenizer.fit_on_texts(quotes)
```

```python
# Retrieve the word-to-integer mapping (word_index) from the tokenizer
word_index = tokenizer.word_index
# Print the total number of unique words found by the tokenizer
print(len(word_index))
# Display the first 10 items in the word_index to see the mapping
list(word_index.items())[:10]
```

```python
# Convert each quote text into a sequence of integer IDs using the fitted tokenizer
sequence = tokenizer.texts_to_sequences(quotes)
```

```python
# Demonstrate the effect of tokenization by comparing the original text with its numerical sequence
##we can clearly see comparison of how tokenization changes sentence
print(quotes[0])
print(sequence[0])
```

```python
x = []
y = []

for seq in sequence:
  for i in range(1, len(seq)):
    x.append(seq[:i])
    y.append(seq[i])
```

```python
# Print the total number of (input, target) pairs generated for training
print(len(x),len(y))
```

## Padding Sequences

Now that we have our `x` (input sequences) and `y` (target words), we need to ensure all input sequences have the same length. This is crucial for feeding them into a neural network. We will use `pad_sequences` to achieve this.

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Determine the maximum sequence length from the generated x sequences
max_sequence_length = max(len(seq) for seq in x)

# Using 'pre' padding is critical for RNNs/LSTMs with long sequences.
# It ensures the actual word data is at the end of the sequence,
# closer to the prediction point, preventing the 'vanishing' of information.
padded_x = pad_sequences(x, maxlen=max_sequence_length, padding='pre')

print(f"Original number of input sequences: {len(x)}")
print(f"Maximum length of an input sequence: {max_sequence_length}")
print(f"Shape of padded input sequences (padded_x): {padded_x.shape}")

# Display the first padded sequence as an example
print("\nFirst padded sequence (padded_x[0]):")
print(padded_x[0])
```

```python
print(x[0],"after padding ->",padded_x[0])
```

```python
y = np.array(y)
print(padded_x.shape , y.shape)
```

```python
from tensorflow.keras.utils import to_categorical
# y_one_hot is created by converting integer labels into binary class matrices
y_one_hot = to_categorical(y, num_classes=vocab_size)

# 'Sequential' allows us to build a model layer-by-layer in a step-by-step fashion.
# It is used for a plain stack of layers where each layer has exactly one input and one output.
```

```python
y_one_hot.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,SimpleRNN,Dense,Embedding
```

```python
embedding_dim = 50
rnn_units = 128
```

```python
rnn_model = Sequential()

# The Embedding layer turns integer IDs into dense vectors of fixed size (embedding_dim).
# It helps the model understand relationships between words in a continuous space.
rnn_model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_sequence_length))

# SimpleRNN is a basic recurrent layer where the output from the previous step is fed back into the current step.
rnn_model.add(SimpleRNN(units=rnn_units))

# The Dense layer with 'softmax' activation acts as the classifier.
# It outputs a probability distribution across the entire vocabulary for the next word.
rnn_model.add(Dense(units=vocab_size, activation='softmax'))
```

```python
# Compile the RNN model
# 'adam' is a popular optimizer that adjusts the learning rate during training.
# 'categorical_crossentropy' is the standard loss function for multi-class classification tasks.
rnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
rnn_model.summary()
```

```python
## Building the LSTM (Long Short-Term Memory) model
lstm_model = Sequential()

## Embedding layer to represent words as dense vectors
lstm_model.add(
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_sequence_length)
)

## LSTM layer: A more advanced recurrent layer capable of learning long-term dependencies.
## It solves the vanishing gradient problem found in SimpleRNN.
lstm_model.add(
    LSTM(units=rnn_units)
)

## Output layer: Predicting the next word from the vocabulary using softmax
lstm_model.add(
    Dense(units=vocab_size, activation='softmax')
)
```

```python
## Compiling the LSTM model with Adam optimizer and Cross-Entropy loss
lstm_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Display the architecture and parameter counts of the LSTM model
lstm_model.summary()
```

```python
epochs = 100
batch_size = 128
hisotry_rnn = lstm_model.fit(padded_x, y_one_hot, epochs=epochs, batch_size=batch_size,validation_split=0.1)
```

```python
lstm_model.save("lstm_model.h5")
```

```python
index_to_word = {}
for word,index in word_index.items():
  index_to_word[index] = word
```

```python
def predictor(model,tokenizer,text,max_sequence_length):
    text = text.lower()

    sequence = tokenizer.texts_to_sequences([text])

    sequence = pad_sequences(sequence,maxlen=max_sequence_length,padding='pre')

    prediction = model.predict(sequence,verbose = 0)
    pred_index = np.argmax(prediction)
    return index_to_word[pred_index]
```

```python
seed_text = "life is"
next_word = predictor(lstm_model,tokenizer,seed_text,max_sequence_length)
print(f"Seed Text: {seed_text}")
print(f"Next Word: {next_word}")
```

```python

def generate_text(model,tokenizer,seed_text,num_words_to_generate,max_sequence_length):
  for _ in range(num_words_to_generate):

    next_word = predictor(model,tokenizer,seed_text,max_sequence_length)
    if next_word == "":
      break
    seed_text += " " + next_word

  return seed_text
```
